# Understandind Decorators

```python
import torch
import time

def mat_mul(arr1, arr2):
    res = torch.matmul(arr1, arr2)
    return res

def timer(func):
    def wrapper(arr1, arr2):
        start = time.time()
        result = func(arr1, arr2)
        end = time.time()
        print(f"The time taken byu the function: {end-start}")
        return result
    return wrapper

arr1 = torch.rand(3, 4)
arr2 = torch.rand(4, 2)

mat_mul = timer(mat_mul)
```

```python
import torch
import time


def timer(func):
    def wrapper(arr1, arr2):
        start = time.time()
        result = func(arr1, arr2)
        end = time.time()
        print(f"The time taken byu the function: {end-start}")
        return result
    return wrapper

@timer
def mat_mul(arr1, arr2):
    res = torch.matmul(arr1, arr2)
    return res

arr1 = torch.rand(3, 4)
arr2 = torch.rand(4, 2)
```

```python
mat_mul(arr1, arr2)
```

# Static method

```python
class Employee():

    def __init__(self, name, id):
        self.name = name
        self.id = id

    def email_id(self):
        email = f"{name}.{id}@company.com"
        return email
    
    @staticmethod
    def conv_id(id):
        return id * 50
```

```python
emp1 = Employee('Akshay', 10)
emp1
```

```python
emp1.conv_id(emp1.id)
```

# Class Method

Class method is a method that has access to the class itself. It recieves cls as its first argument, which refers to the class.

## Concept 1: Python allows only one  __init__. But, what if we wanna create objects in different way ?

```python
class student:

    def __init__(self, name, grade, age):
        self.name = name
        self.grade = grade
        self.age = age

    @classmethod
    def from_dict(cls, data):
        return cls(data['name'], data['grade'], data['age'])
    
    @classmethod
    def from_str(cls, data):
        lst = data.split('-')
        return cls(lst[0], lst[1], lst[2])
```

```python
s1 = student('Akshay', 9, 28)
s2 = student.from_dict({'name': 'Rithik', 'grade': 6, 'age': 34})
s3 = student.from_str('Akash-8-34')
```
