```python
import torch
import time


def timer(func):
    def wrapper(arr1, arr2):
        start = time.time()
        result = func(arr1, arr2)
        end = time.time()
        print(f"The time taken byu the function: {end-start}")
        return result
    return wrapper

@timer
def mat_mul(arr1, arr2):
    res = torch.matmul(arr1, arr2)
    return res

arr1 = torch.rand(3, 4)
arr2 = torch.rand(4, 2)
```

```python
mat_mul(arr1, arr2)
```

# Class Method

Class method is a method that has access to the class itself. It recieves cls as its first argument, which refers to the class.

## Concept 1: Python allows only one  __init__. But, what if we wanna create objects in different way ?

```python
class student:

    def __init__(self, name, grade, age):
        self.name = name
        self.grade = grade
        self.age = age

    @classmethod
    def from_dict(cls, data):
        return cls(data['name'], data['grade'], data['age'])
    
    @classmethod
    def from_str(cls, data):
        lst = data.split('-')
        return cls(lst[0], lst[1], lst[2])
```

```python
s1 = student('Akshay', 9, 28)
s2 = student.from_dict({'name': 'Rithik', 'grade': 6, 'age': 34})
s3 = student.from_str('Akash-8-34')
```

```python
print(s1.name, s1.grade, s1.age)
print(s2.name, s2.grade, s2.age)
print(s3.name, s3.grade, s3.age)
```

```python
## Realworld example:

from datetime import datetime

print(datetime(2024, 1, 5))
print(datetime.now())
print(datetime.fromtimestamp(1700000000))
print(datetime.fromisoformat("2024-01-15"))
```

## Concept 2: Classmethods also come handy in changing the class variable

```python
class employee:

    comp_name = 'Michelin'
    emp_count = 0

    def __init__(self, name, age):
        self.name = name
        self.age = age
        employee.emp_count += 1

    @classmethod
    def get_company_name(cls):
        print('Accessing the variable')
        return cls.comp_name 
    
    @classmethod
    def change_company(cls, new_comp):
        print("Changing the company name")
        cls.comp_name = new_comp

    @classmethod
    def num_employees(cls):
        return cls.emp_count
```

# Encapsulation

Encapsulation means hiding the internal data of an object and only exposing what is necessary. It's about protecting your data from being accidentally changed from outside.

```python
class reportcard():

    def __init__(self, name, marks):
        self.name = name
        self._marks = marks
        self.__grade = None

        self.__calculate_grade()

    # We want this calculate_grade to be operated only within the class and not outside.
    def __calculate_grade(self):
        if self._marks >= 90:
            self.__grade = "A"
        elif self._marks >= 75:
            self.__grade = "B"
        elif self._marks >= 60:
            self.__grade = "C"
        else:
            self.__grade = "F"

    @property # In place of get_grade, we can use @property
    def grade(self):
        return self.__grade

    @property # In place of get_marks, we can use @property
    def marks(self):
        return self._marks
    
    @marks.setter
    def marks(self, new_marks):
        if not (0 <= new_marks <= 100):
            raise ValueError("Marks must be between 0 and 100!")
        self._marks = new_marks
        self.__calculate_grade()
```

```python
s1 = reportcard('Akshay', 95)
```

```python
print(s1.name)
print(s1.marks)
print(s1.grade)
```

```python
s1.marks = 52
```

```python
print(s1.grade)
```

# Inheritance

Inheritance is the way in which child class can aquire the properties of the parent class

## Type 1: Single Inheritance

```python
class Animal:
    def __init__(self, name):
        self.name = name

    # Common Features
    def features(self):
        print(f"{self.name} has two eyes")
        print(f"{self.name} has a tail")
        print(f"{self.name} has two ears")

class Dog(Animal):
    def sound(self):
        print(f"{self.name} says WOOF!")

class cat(Animal):
    def sound(self):
        print(f"{self.name} says MEOW!")
```

```python
a1 = Dog("Pinki")
print(a1.name)
print(a1.features())
print(a1.sound())
```

## Type 2: Multilevel Inheritance

```python
class Father:
    def coding(self):
        print("Father knows coding")

class Mother:
    def cooking(self):
        print("Mother knows cooking")

class Child(Father, Mother):    # inherits from both!
    def drawing(self):
        print("Child knows drawing")


c = Child()
c.coding()      # from Father
c.cooking()     # from Mother
c.drawing()     # Child's own
```

## Type 3: Method Overriding

Sometime we want to extend the parent's method, not completely replace it

```python
class animal:
    def speak(self):
        print('Animal is about to speak')

class Dog(animal):
    def speak(self):
        super().speak()             # run parent's version first
        print("Animal says WOOF")   # then add extra behaviour
```

```python
d1 = Dog()
```

```python
d1.speak()
```

## Using Super()

super() helps us to call the parent's method from inside the child class

```python
# Without super()
class animal:
    def __init__(self, name, age):
        self.name = name
        self.age = age

class Dog(animal):
    def __init__(self, name, age, breed):
        self.breed = breed

a1 = Dog('Bruno', 5, 'Labrador')
```

```python
print(a1.name)
print(a1.age)
```

When the Dog class made its own __ini__ as def __init__(self, name, age, breed), it COMPLETELY REPLACES the animal's __init__. So, self.name and self.age never get set unless we EXPLICITY CALL the animal's __init__.

```python
# With the use of super()
class animal:
    def __init__(self, name, age):
        self.name = name
        self.age = age

class Dog(animal):
    def __init__(self, name, age, breed):
        super().__init__(name, age) # Manually Trigerring animal's init
        self.breed = breed

a1 = Dog('Bruno', 5, 'Labrador')
```

```python
a1.name, a1.age, a1.breed
```

## Practice

```python
class employee:
    def __init__(self, name, salary):
        self.name = name
        self.__salary = salary

    @property
    def salary(self):
        return self.__salary

    def work(self):
        print(f"{self.name} is working.")
    
    def show_details(self):
        print(f"{self.name} is working in a company with the salary {self.__salary}")

class manager(employee):
    def __init__(self, name, age, salary, team_size):
        super().__init__(name, salary)
        self.age = age
        self.team_size = team_size

    def manage(self):
        print(f"{self.name} manages the team")

    def show_details(self):
        super().show_details()
        print(f"{self.name} is managing the team of {self.team_size}")

class intern(employee):
    def __init__(self, name, age, salary, duration):
        super().__init__(name, salary)
        self.duration = duration
        self.age = age

    def show_details(self):
        super().show_details()
        print(f"{self.name} is interning for {self.duration} months")
```

```python
m1 = intern('Akshay', 25, 50000, 8)
```

```python
m1.name, m1.age, m1.salary, m1.duration
```

```python
m1.show_details()
```
